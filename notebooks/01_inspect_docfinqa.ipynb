```python
from huggingface_hub import hf_hub_download

train_path = hf_hub_download(
    repo_id="kensho/DocFinQA",
    filename="train.json",
    repo_type="dataset"
)

print(train_path)
```

```python
import ijson

with open(train_path, "rb") as f:
    first_example = next(ijson.items(f, "item"))

print(first_example.keys())
```

```python
print("QUESTION:")
print(first_example["Question"])

print("\nANSWER:")
print(first_example["Answer"])

print("\nPROGRAM:")
print(first_example["Program"])

print("\nCONTEXT LENGTH:")
print(len(first_example["Context"]))
```

```python
print(first_example["Context"][:3000])
```

```python
for key, value in first_example.items():
    print(key, type(value), len(str(value)))
```

```python
clean_example = {
    "document_text": first_example["Context"],
    "question": first_example["Question"],
    "program": first_example["Program"],
    "gold_answer": first_example["Answer"],
}

print(clean_example.keys())
print(clean_example["question"])
print(clean_example["gold_answer"])
```

```python
def load_first_docfinqa_example(path):
    with open(path, "rb") as f:
        raw_example = next(ijson.items(f, "item"))

    return {
        "document_text": raw_example["Context"],
        "question": raw_example["Question"],
        "program": raw_example["Program"],
        "gold_answer": raw_example["Answer"],
    }


example = load_first_docfinqa_example(train_path)

print("Question:", example["question"])
print("Gold Answer:", example["gold_answer"])
print("Document length:", len(example["document_text"]))
```
